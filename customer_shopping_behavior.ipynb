```python
import pandas as pd
df = pd.read_csv('/content/customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df =df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
#membuat kelompok umur
labels = ['Young Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age group'] = pd.cut(df['age'], bins=4, labels=labels)
```

```python
df[['age','age group']].head(10)
```

```python
#membuat kolom frekuensi pembayaran setiap hari nya
frequency_mapping= {
    'fortnightly': 14,
    'weekly': 7,
    'monthly': 30,
    'quarterly': 90,
    'bi-weekly' : 14,
    'annually': 365,
    'every 3 months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].str.lower().map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis= 1)
```
